```python
sr = 44100 #Samplingfrequenz 44.1kHz
T = 1  #Anzahl Sekunden
N = sr*T  #Samplingfrequenz * Sekunden 

freqAxis = linspace(0,sr/2,N//2)  #Frequenzachse - Array bildet den Wertebereich von 0 bis 22050 in N/2 Samples ab.
```

```python
y, sr = sf.read('test_click_apE_1_-0.8.wav')  #Importieren der Impulsantwort

figure(figsize=[20,6])
subplot(121)
plot(y) #Plottet die Wellenform der Impulsantwort 
plt.xlabel('Time (Samples)')
plt.ylabel('Amplitude', labelpad=-5)
plt.xlim(0,30) #Limit the x-Axis
grid()

title('$h(n)$')
```

```python
spec = fft.fft(y)[0:N//2] 

figure(figsize=[20,6])

subplot(121)
plt.axis([min(freqAxis), max(freqAxis), 0.0, 2])
plot(freqAxis,abs(spec))
plt.xlabel('Frequency')
plt.ylabel('Magnitude', labelpad=5)

title('$Magnitude$ $Response$')
grid()


subplot(122)
plot(freqAxis,angle(spec, deg = True))
plt.xlabel('Frequency')
plt.ylabel('Degree')

title('$Phase$ $Response$')
grid()

plt.subplots_adjust(left=0.1,
                    bottom=0.1, 
                    right=0.9, 
                    top=0.9, 
                    wspace=0.4, 
                    hspace=0.4)
```
